```python
import requests

try:
    response = requests.get("http://localhost:11434/api/tags", timeout=0.1)
    if response.status_code == 200:
        print("Ollama 서버가 실행 중입니다.")
    else:
        print(f"Ollama 서버 응답 코드: {response.status_code}")
except requests.exceptions.RequestException as e:
    print("Ollama 서버가 꺼져 있거나 연결할 수 없습니다.", e)
```
